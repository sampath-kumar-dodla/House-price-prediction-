```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
data=pd.read_csv('Delhi house data.csv')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
data.pop('Locality')
```

```python
data
```

```python
data.info()
```

```python
data = data.join(pd.get_dummies(data['Status'], prefix='Status') * 1).drop(['Status'], axis=1)
data = data.join(pd.get_dummies(data['Transaction'], prefix='Transaction') * 1).drop(['Transaction'], axis=1)
data = data.join(pd.get_dummies(data['Type'], prefix='Type') * 1).drop(['Type'], axis=1)
data = data.join(pd.get_dummies(data['Furnishing'], prefix='Furnishing') * 1).drop(['Furnishing'], axis=1)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = data.drop(['Price'], axis=1)
y = data['Price']
reg = LinearRegression()
reg.fit(x, y)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
train_data = x_train.join(y_train)
test_data = x_test.join(y_test)
```

```python
reg.score(x_test, y_test)
```
